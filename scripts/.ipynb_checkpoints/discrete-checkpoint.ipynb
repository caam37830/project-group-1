```python
# Phase Diagram Code
print('I am running')

Ts = [10, 50, 100, 300, 600]

for t in Ts:
    # bs = np.linspace(0.001, 0.01, 10)
    # ks = np.linspace(0.1, 0.6, 10)
    bs = np.logspace(-4, 0, 10)
    ks = np.logspace(-4, 0, 10)
    cts = np.zeros((len(bs), len(ks)))
    for i, b in enumerate(bs):
        for j, k in enumerate(ks):
            cts[i,j] = run_simulation_phase(b, k, T = t)
    plt.figure(figsize=(10,5))
    plt.imshow(cts, extent=[np.min(bs), np.max(bs), np.min(ks), np.max(ks)])
    plt.colorbar()
    # plt.axis('square')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('b')
    plt.ylabel('k')
    plt.show()
```
